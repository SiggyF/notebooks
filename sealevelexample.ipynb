```python
%matplotlib inline
import pandas
import requests
import io
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'http://www.psmsl.org/data/obtaining/rlr.monthly.data/32.rlrdata'
```

```python
print requests.get(url).content[:100]
```

```python
# in memory file
f = io.BytesIO(requests.get(url).content)
df = pandas.read_csv(f, sep=';', names=('year', 'waterlevel', 'code', 'another code'))
df.head()
```

```python
fig, ax = plt.subplots()
plt.plot(df['year'], df['waterlevel'])
ax.set_ylabel('water level [mm]')
ax.set_xlabel('year')
```

```python
import statsmodels.api as sm
```

```python
y = df['waterlevel']
X = df['year']
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
def fit(startyear):
    df_selected = df[df['year'] > startyear]
    y = df_selected['waterlevel']
    X = df_selected['year']
    X = sm.add_constant(X)
    model = sm.OLS(y, X)
    results = model.fit()
    print(results.summary())
```

```python
from IPython.html.widgets import interact
```

```python
interact(fit, startyear=(1860, 1980, 10))
```

```python
def plot(startyear):
    df_selected = df[df['year'] > startyear]
    y = df_selected['waterlevel']
    X = df_selected['year']
    X = sm.add_constant(X)
    model = sm.OLS(y, X)
    results = model.fit()
    fig, ax = plt.subplots()
    ax.plot(df['year'], df['waterlevel'], '.', alpha=0.1)
    ax.plot(df_selected['year'], results.fittedvalues)
    
    ax.set_ylim(6400,7300)
```

```python
interact(plot, startyear=(1860, 1980, 10))
```

```python
from statsmodels.sandbox.regression.predstd import wls_prediction_std


def plot(startyear):
    df_selected = df[df['year'] > startyear]
    y = df_selected['waterlevel']
    X = df_selected['year']
    X = sm.add_constant(X)
    model = sm.OLS(y, X)
    results = model.fit()
    fig, ax = plt.subplots()
    ax.plot(df['year'], df['waterlevel'], '.', alpha=0.1)
    ax.plot(df_selected['year'], results.fittedvalues)
    fit, lower, upper = wls_prediction_std(results)
    plt.fill_between(df_selected['year'], lower, upper, alpha=0.3)
    
    ax.set_ylim(6400,7300)
```

```python
interact(plot, startyear=(1860, 1980, 10))
```

```python
def plot(startyear):
    n_periods = 5
    df_selected = df[(df['year'] >= startyear) & (df['year'] < (startyear + 18.6*n_periods))]
    
    y = df_selected['waterlevel']
    X = np.c_[
        df_selected['year']-1970, 
        np.cos(2*np.pi*(df_selected['year']-1970)/18.613),
        np.sin(2*np.pi*(df_selected['year']-1970)/18.613)

    ]
    X = sm.add_constant(X)
    model = sm.OLS(y, X)
    results = model.fit()
    # e^{i\theta} = \cos\theta + i\sin\theta
    phase = np.angle(complex(results.params['x2'], results.params['x3']))
    ampl = np.sqrt(results.params['x2']**2 + results.params['x3']**2)
    
    # plot it
    fig, ax = plt.subplots()
    ax.set_title("Phase (rad): %.1f, amplitude: %.1f\nA(cos): %.1f, B(sin): %.1f" % (phase, ampl, results.params['x2'], results.params['x3']))
    ax.plot(df['year'], df['waterlevel'], '.', alpha=0.3)
    ax.plot(df_selected['year'], results.fittedvalues, linewidth=5)
    
```

```python
interact(plot, startyear=(1860, 1980, 5))
```

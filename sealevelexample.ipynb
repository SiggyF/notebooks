```python
import pandas
import requests
import io
import matplotlib.pyplot as plt
```

```python
url = 'http://www.psmsl.org/data/obtaining/rlr.monthly.data/23.rlrdata'
```

```python
print requests.get(url).content[:100]
```

```python
# in memory file
f = io.BytesIO(requests.get(url).content)
df = pandas.read_csv(f, sep=';', names=('year', 'waterlevel', 'code', 'another code'))
df.head()
```

```python
fig, ax = plt.subplots()
plt.plot(df['year'], df['waterlevel'])
ax.set_ylabel('water level [mm]')
ax.set_xlabel('year')
```

```python
import statsmodels.api as sm
```

```python
y = df['waterlevel']
X = df['year']
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
def fit(startyear):
    df_selected = df[df['year'] > startyear]
    y = df_selected['waterlevel']
    X = df_selected['year']
    X = sm.add_constant(X)
    model = sm.OLS(y, X)
    results = model.fit()
    print(results.summary())
```

```python
from IPython.html.widgets import interact
```

```python
interact(fit, startyear=(1860, 1980, 10))
```

```python
def plot(startyear):
    df_selected = df[df['year'] > startyear]
    y = df_selected['waterlevel']
    X = df_selected['year']
    X = sm.add_constant(X)
    model = sm.OLS(y, X)
    results = model.fit()
    fig, ax = plt.subplots()
    ax.plot(df['year'], df['waterlevel'], '.', alpha=0.1)
    ax.plot(df_selected['year'], results.fittedvalues)
    
    ax.set_ylim(6400,7300)
```

```python
interact(plot, startyear=(1860, 1980, 10))
```

```python
from statsmodels.sandbox.regression.predstd import wls_prediction_std


def plot(startyear):
    df_selected = df[df['year'] > startyear]
    y = df_selected['waterlevel']
    X = df_selected['year']
    X = sm.add_constant(X)
    model = sm.OLS(y, X)
    results = model.fit()
    fig, ax = plt.subplots()
    ax.plot(df['year'], df['waterlevel'], '.', alpha=0.1)
    ax.plot(df_selected['year'], results.fittedvalues)
    fit, lower, upper = wls_prediction_std(results)
    plt.fill_between(df_selected['year'], lower, upper, alpha=0.3)
    
    ax.set_ylim(6400,7300)
```

```python
interact(plot, startyear=(1860, 1980, 10))
```

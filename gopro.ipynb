```python
#define the GCP points in pixels, srcpoints, and the GCP points in RD coordinates, dstpoints
srcpoints  = np.array([[1312,1048],[2794,796],[1441,1877],[2836,1826],[2056,1440]]).astype(np.float32)
dstpoints  = np.array([[85670.316,445806.969],[85700.903,445821.268],[85678.655,445788.205],[85709.590,445799.925],[85689.973,445803.060]]).astype(np.float32)
dstpoints2 = np.array([20 * (i - [85630,445705.66]) for i in dstpoints]).astype(np.float32)
#find the transformation matrix
M,  dc  = cv2.findHomography(srcpoints,dstpoints)
M2, dc2 = cv2.findHomography(srcpoints,dstpoints2)
M       = M.astype(float32)
#set up the new coordinate systems from pixels to RD coordinates
X,Y = meshgrid(np.arange(testimg.shape[1]),
               np.arange(testimg.shape[0]))
#XY  = np.c_[X.flatten(),Y.flatten()][:,:, np.newaxis].astype('float32')
#XY = np.tile(XY, [1,1,2])
XY = np.concatenate([X[:,:,np.newaxis],Y[:,:,np.newaxis]], axis=2).astype('float32')
RD  = cv2.perspectiveTransform(XY,M)
RDx, RDy = RD[...,0], RD[...,1]

#RDx = RD[:,0].reshape(testimg.shape[:2])
#RDy = RD[:,1].reshape(testimg.shape[:2])
```

```python
imggeo            = cv2.warpPerspective(testimg,M2,(2500,3000)) # rectify the original test image
imggeo[imggeo==0] = 255 # set the black background to white
d = 10 #number of steps in constructing the pcolormesh

fig2, ax = subplots(1,2,figsize=(30,15))
fig2.tight_layout

ax[0].imshow(imggeo[:,:,:],origin='lower')
ax[0].set_xlabel('x RD-coordinate [m]', fontsize=14)
ax[0].set_ylabel('y RD-coordinate [m]', fontsize=14)
ax[0].tick_params(labelsize=14)
ax[0].set_xticks((arange(85640,85780,20)-85630)*20)
ax[0].set_yticks((arange(20,160,20)+4.457e5-445705.66)*20)
xt = ax[0].get_xticks().tolist()
yt = ax[0].get_yticks().tolist()
ax[0].set_xticklabels((asarray(xt)/20)+85630)
ax[0].set_yticklabels((asarray(yt)/20)+445705.66-4.457e5)

ax[1].pcolormesh(RDx[::d,::d],RDy[::d,::d],testimg[::d,::d,0].astype('float64')/255,cmap=cm.Greys_r)
ax[1].set_xlim([85630,85755])
ax[1].set_ylim([5.65999999997+4.457e5,155.66+4.457e5])
ax[1].set_xlabel('x RD-coordinate [m]', fontsize=14)
ax[1].set_ylabel('y RD-coordinate [m]', fontsize=14)
ax[1].tick_params(labelsize=14)
```
